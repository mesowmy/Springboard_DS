```python
def avg_price_ratio(pr):
    avg_price_2003 = pr[pr['Year'] == 2003] ['Average_price'].mean()
    avg_price_2023 = pr[pr['Year'] == 2023] ['Average_price'].mean()
    #check for missing data
    if avg_price_2023 == 0 or pd.isna(avg_price_2003) or pd.isna(avg_price_2023):
        return None #or some value
    return avg_price_2003/avg_price_2023
price_ratio = {}
#iterate
for borough in dfg['London_Borough'].unique(): 
#filter applied to create a new data frame with data for that borough
    unique_borough = dfg[dfg['London_Borough'] == borough]
    price_ratio[borough] = avg_price_ratio(unique_borough) # store in the directory, final_ratio

print(price_ratio)
```

```python
#convert dictionary to df
price_ratio_df = pd.DataFrame(price_ratio.items(), columns = ['London_Borough', 'price_ratio'])
print(price_ratio_df)
```

```python
#Sort the df, from the ones that saw the highest-lowest change, low to higher ratio. in place will modify and save as the same df
price_ratio_df.sort_values(by = 'price_ratio', ascending = True, inplace = True)
print(price_ratio_df)
```
